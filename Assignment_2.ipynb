```python
try:
    with open("data.txt", "r") as file:
        lines = file.readlines()

    word_count = sum(len(line.split()) for line in lines)
    line_count = len(lines)

except FileNotFoundError:
    print("The file was not found.")

except PermissionError:
    print("No permission to access the file.")

else:
    print("Number of words:", word_count)
    print("Number of lines:", line_count)
```

### Task 5

```python
import re

email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'

with open("log.txt", "r") as log:
    log_text = log.read()

emails = re.findall(email_pattern, log_text)
print(emails)
```

```python
phone_extract_pattern = r'(?:\+1[\s-]?)?(?:\(\d{3}\)|\d{3})[\s.-]?\d{3}[\s.-]?\d{4}'
phone_validate_pattern = r'^(?:\+1[ -]?)?(?:\(\d{3}\)|\d{3})[ .-]?\d{3}[ .-]?\d{4}$'
```

```python
with open("phone_numbers.txt", "r") as pn:
    phone_text = pn.read()
```

```python
phone_numbers = re.findall(phone_extract_pattern, phone_text)
```

```python
valid_phones = []
for number in phone_numbers:
    if re.fullmatch(phone_validate_pattern, number):
        valid_phones.append(number)

print(valid_phones)
```

#### Fixed Code:

```python
import math
import logging
import math_operations
```

```python
logging.basicConfig(level=logging.INFO)
```

```python
def calculate_area(radius):
    return math.pi * radius * radius

try:
    radius = input("Enter the radius: ")
    logging.info("Input radius: %s", radius)

    radius = float(radius)
    area = calculate_area(radius)

    logging.info("Area: %s", area)
    print("Area:", area)

except ValueError:
    logging.error("Invalid input.")
    print("Error: Please enter a valid number.")
```
